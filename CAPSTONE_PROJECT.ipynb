```python
# Function to detect outliers using IQR method
def detect_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outlier_mask = (df[col] < lower_bound) | (df[col] > upper_bound)
        if outlier_mask.any():
            outliers[col] = df[col][outlier_mask]
    return outliers

# List of renamed numerical columns in the DataFrame 'data_sub'
r_columns = ['VOICE_REVNUE_IN_30DAYS', 'DATA_USED_IN_30DAYS', 'TOTAL_DATA_IN_30DAYS',
                   'VOICE_USAGE_IN_30DAYS', '4G_DATA_USED_IN_30DAYS']

# Detect outliers in the renamed columns
outliers = detect_outliers_iqr(data, r_columns)
```

## Hypothesis Testing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
# 1.One-sample t-test (Testing against a population mean):
# Compute the mean of the variable
mean_revenue = data['REVNUE_GENRTD_IN_LAST_30DAYS'].mean()

# Perform the one-sample t-test
alpha = 0.01
t_statistic, p_value = st.ttest_1samp(a=data['REVNUE_GENRTD_IN_LAST_30DAYS'], popmean=mean_revenue)
print("One-sample t-test Results:")
print(f"t-statistic: {t_statistic}, p-value: {p_value}")
if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')
```

```python
# 2.Independent samples t-test (Testing between two groups):
data_m = data[data['SMART_PHN_FLAG'] == 'N']['REVNUE_GENRTD_IN_LAST_30DAYS']
data_f = data[data['SMART_PHN_FLAG'] == 'Y']['REVNUE_GENRTD_IN_LAST_30DAYS']

# Perform the independent samples t-test
alpha = 0.01
t_statistic, p_value = st.ttest_ind(data_m, data_f)
print("Independent Samples T-test Results:")
print(f"t-statistic: {t_statistic}, p-value: {p_value}")
if p_value < alpha:
    print('Reject the null hypothesis')
else:
    print('Accept the null hypothesis')
```

### Chi Square

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Select a small subset of the data
subset_data = data.iloc[:1000] #couldnt be done on full set because of memory error

# Select a few categorical features for the chi-square test
categorical_features = ['DEVICE_TYPE', 'CUST_SEGMNT', 'DEVICE_NETWORK']

# Perform the chi-square test for each selected feature
results = []
for feature in categorical_features:
    # Create a contingency table between the feature and 'REVNUE_GENRTD_IN_LAST_30DAYS'
    table = pd.crosstab(subset_data[feature], subset_data['REVNUE_GENRTD_IN_LAST_30DAYS'])
    
    # Perform the chi-square test for independence
    stat, p, dof, expected = chi2_contingency(table)
    
    # Append the results to the list
    results.append((feature, p))

# Create a DataFrame to display the results
result_df = pd.DataFrame(results, columns=['Feature', 'P-value'])
result_df.sort_values('P-value', inplace=True)

print(result_df)

```
